```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
```

```python
df = pd.DataFrame({
    "job" : ["tech" , "admin" , "manager"],
    "age" : [30,40,10],
    "sick" : ["no" , "yes" , "yes"]
})
df
label_encoder = LabelEncoder()
df['job']= label_encoder.fit_transform(df['job']) 
df['sick']= label_encoder.fit_transform(df['sick']) 
df
```

```python
X = df.iloc[:,:2]
y = df.iloc[:,2]
```

```python
dt = tree.DecisionTreeClassifier(random_state=102)
dt.fit(X,y)
```

```python
dt.predict([[0,20] , [1,10]])
```

```python
tree.plot_tree(dt)
```
